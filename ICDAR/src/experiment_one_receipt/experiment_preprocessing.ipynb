# Experiment Preprocessing

## Import

```python
import pytesseract
import cv2
from matplotlib import pyplot as plt
from functions.image_preprocessing import image_preprocessing
```

## Leitura de imagens

```python
image_original = cv2.imread('../../data/input/receipt.jpg', cv2.IMREAD_GRAYSCALE)
```

## Pr√©-processamento

```python
image_preprocessed = image_preprocessing(image_original)

fig = plt.figure(figsize=(60,35))
fig.add_subplot(1,3,1)
plt.imshow(image_original, cmap="gray")
plt.title('Original')
plt.axis('off')

fig.add_subplot(1,3,2)
plt.imshow(image_preprocessed, cmap="gray")
plt.title('Binary')
plt.axis('off')

plt.show()
```

## Extra√ß√£o de dados

```python
text = pytesseract.image_to_string(image_preprocessed, lang='eng').upper()
print(text)
```
